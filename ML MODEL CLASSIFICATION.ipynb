```python
x = df[['mass','width','height','color_score']]
y = df[['fruit_name']]
# divide dataframe into x-input columns and y-output columns
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.15,random_state=13)
```

```python
x.shape
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=5,metric='euclidean')
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
#import all the classification algorithms to compare accuracies
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# LogisticRegression
clf = LogisticRegression(random_state=0)
clf.fit(x_train, y_train)
# Prediction
y_pred = clf.predict(x_test)

acc=accuracy_score(y_test,y_pred)
print("Logistic Regression Model accuracy (in %):", acc*100)
```

```python
# the accuracy of Logistic Regression Model is 66.66%
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

clf = GaussianNB()
clf.fit(x_train, y_train)
# Prediction
y_pred = clf.predict(x_test)

acc = accuracy_score(y_test, y_pred)
print(" GaussianNB (in %):", acc*100)
```

```python
# the accuracy of GaussianNB is 77.77%
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score

clf = SGDClassifier(random_state=0)
clf.fit(x_train, y_train)
# Prediction
y_pred = clf.predict(x_test)

acc = accuracy_score(y_test, y_pred)
print(" SGDClassifier model accuracy (in %):", acc*100)
```

```python
# the accuracy of SGDClassifier model is 11.11%
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

clf = KNeighborsClassifier()
clf.fit(x_train, y_train)
# Prediction
y_pred = clf.predict(x_test)

acc = accuracy_score(y_test, y_pred)
print("KNeighborsClassifier model accuracy (in %):", acc*100)
```

```python
# the accuracy of KNeighborsClassifier is 55.55%
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# LogisticRegression
clf = DecisionTreeClassifier()
clf.fit(x_train, y_train)
# Prediction
y_pred = clf.predict(x_test)

acc = accuracy_score(y_test, y_pred)
print("Decision Tree Classifier model accuracy (in %):", acc*100)
```

```python
# the accuracy of DecisionTreeClassifier is 100%
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# LogisticRegression
clf = SVC()
clf.fit(x_train, y_train)
# Prediction
y_pred = clf.predict(x_test)

acc = accuracy_score(y_test, y_pred)
print("support vector machine SVC model accuracy (in %):", acc*100)
```

```python
# the accuracy of SVC model is 33.33%
```
